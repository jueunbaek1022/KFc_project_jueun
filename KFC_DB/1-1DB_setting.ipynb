```python
import pymysql 
```

## 1. DB연결
- DB 관리자 이외의 팀원들이 DB에 함께 들어오기 위한 코드
- DB 관리자로부터 아래 정보 받기

```python
# 없다면 설치 필요: pip install python-dotenv
```

```python
import os
from dotenv import load_dotenv

# .env 파일 경로 지정 
load_dotenv()

# 키 설정하기
HOST = os.getenv('HOST')
PORT = int(os.getenv('PORT'))   
USER = os.getenv('USER')
PASSWD = os.getenv('PASSWD')
DB = os.getenv('DB')
```

```python
conn = pymysql.connect(host=HOST, port=PORT, user=USER, passwd=PASSWD, db=DB, charset='utf8')
```

## 2. DB cursor 생성

```python
cursor = conn.cursor()
```

## 3. 예제 쿼리 실행(기존 데이터 조회 예시)

```python
cursor.execute("SELECT * FROM cabbage")
rows = cursor.fetchall()

for row in rows:
    print(row)
```

## 4. 연결 종료

```python
cursor.close()
conn.close()
```
