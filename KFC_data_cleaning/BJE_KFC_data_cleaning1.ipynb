```python
import pandas as pd
```

```python
# CSV 파일 읽기 (경로를 본인의 파일 경로로 수정)
df = pd.read_csv('price/6_green_onions_new.csv', header=[0,1], index_col=0)
```

```python
df
```

```python
# 열 이름에 있는 공백 제거
df.columns = df.columns.map(lambda x: (x[0], x[1].strip()))
```

```python
# 데이터를 정리할 리스트 초기화
data_list = []
```

```python
# 각 월과 연도별로 'Early', 'Mid', 'Late' 값을 가져와 딕셔너리 형태로 추가
for month in df.index:
    for year in df.columns.levels[0]:
        # 'Early', 'Mid', 'Late' 값이 존재하는지 확인한 후 가져오기
        early = df.loc[month, (year, 'Early')] if (year, 'Early') in df.columns else None
        mid = df.loc[month, (year, 'Mid')] if (year, 'Mid') in df.columns else None
        late = df.loc[month, (year, 'Late')] if (year, 'Late') in df.columns else None

        data_list.append({
            'year': year,
            'Month': month,
            'Early': early,
            'Mid': mid,
            'Late': late
        })
```

```python
# 리스트를 DataFrame으로 변환
green_onions_new_try = pd.DataFrame(data_list)
```

```python
# 결과 확인
print(green_onions_new_try)
```

```python
# 변환된 데이터 저장
green_onions_new_try.to_csv('green_onions_new_try.csv', index=False)
```
